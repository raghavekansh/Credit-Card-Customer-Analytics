```python
print("="*70)
print("üîß FEATURE ENGINEERING FOR SEGMENTATION")
print("="*70)

# Create a working copy
df_segment = df.copy()

# RFM Components:
# R (Recency) = Months_Inactive_12_mon (lower is better - more recent activity)
# F (Frequency) = Total_Trans_Ct (higher is better)
# M (Monetary) = Total_Trans_Amt (higher is better)

# Additional behavioral features
df_segment['Avg_Transaction_Value'] = df_segment['Total_Trans_Amt'] / df_segment['Total_Trans_Ct']
df_segment['Credit_Utilization_Pct'] = df_segment['Avg_Utilization_Ratio'] * 100
df_segment['Available_Credit_Pct'] = (df_segment['Avg_Open_To_Buy'] / df_segment['Credit_Limit']) * 100

# Engagement Score (composite metric)
df_segment['Engagement_Score'] = (
    df_segment['Total_Trans_Ct'] * 0.4 +  # Frequency weight
    (df_segment['Total_Trans_Amt'] / 100) * 0.4 +  # Monetary weight (scaled)
    (6 - df_segment['Months_Inactive_12_mon']) * 0.2  # Recency weight (inverted)
)

print("\n‚úÖ New features created:")
print("   ‚Ä¢ Avg_Transaction_Value")
print("   ‚Ä¢ Credit_Utilization_Pct")
print("   ‚Ä¢ Available_Credit_Pct")
print("   ‚Ä¢ Engagement_Score")

print("\nüìä Sample of new features:")
display(df_segment[['Total_Trans_Amt', 'Total_Trans_Ct', 'Avg_Transaction_Value',
                     'Engagement_Score', 'Months_Inactive_12_mon']].head(10))

# Summary statistics
print("\nüìà Engagement Score Distribution:")
print(df_segment['Engagement_Score'].describe())
```

```python
print("="*70)
print("üíØ RFM SCORING SYSTEM")
print("="*70)

# Create RFM scores using quartiles (1-4 scale, 4 being best)
# Using rank-based method to handle duplicates better

# Recency Score (invert: lower months inactive = higher score)
df_segment['R_Score'] = pd.cut(df_segment['Months_Inactive_12_mon'],
                                bins=[-1, 1, 2, 3, 7],
                                labels=[4, 3, 2, 1])

# Frequency Score (higher transaction count = higher score)
df_segment['F_Score'] = pd.qcut(df_segment['Total_Trans_Ct'],
                                 q=4, labels=[1, 2, 3, 4], duplicates='drop')

# Monetary Score (higher transaction amount = higher score)
df_segment['M_Score'] = pd.qcut(df_segment['Total_Trans_Amt'],
                                 q=4, labels=[1, 2, 3, 4], duplicates='drop')

# Convert to numeric (handling any NaN values)
df_segment['R_Score'] = pd.to_numeric(df_segment['R_Score'], errors='coerce').fillna(2).astype(int)
df_segment['F_Score'] = pd.to_numeric(df_segment['F_Score'], errors='coerce').fillna(2).astype(int)
df_segment['M_Score'] = pd.to_numeric(df_segment['M_Score'], errors='coerce').fillna(2).astype(int)

# Combined RFM Score
df_segment['RFM_Score'] = (df_segment['R_Score'] * 100 +
                           df_segment['F_Score'] * 10 +
                           df_segment['M_Score'])

# RFM Segment (string representation)
df_segment['RFM_Segment'] = (df_segment['R_Score'].astype(str) +
                              df_segment['F_Score'].astype(str) +
                              df_segment['M_Score'].astype(str))

print("‚úÖ RFM Scores calculated!")
print("\nüìä Sample RFM Analysis:")
display(df_segment[['CLIENTNUM', 'Months_Inactive_12_mon', 'Total_Trans_Ct',
                     'Total_Trans_Amt', 'R_Score', 'F_Score', 'M_Score',
                     'RFM_Score', 'RFM_Segment']].head(10))

# RFM Score distribution
print("\nüìà RFM Score Distribution:")
print(df_segment['RFM_Score'].describe())

# Top RFM segments
print("\nüèÜ TOP 10 RFM SEGMENTS (by customer count):")
top_segments = df_segment['RFM_Segment'].value_counts().head(10)
display(pd.DataFrame({'RFM_Segment': top_segments.index, 'Customer_Count': top_segments.values}))

# Verify scores are created
print(f"\n‚úÖ R_Score range: {df_segment['R_Score'].min()} to {df_segment['R_Score'].max()}")
print(f"‚úÖ F_Score range: {df_segment['F_Score'].min()} to {df_segment['F_Score'].max()}")
print(f"‚úÖ M_Score range: {df_segment['M_Score'].min()} to {df_segment['M_Score'].max()}")
```

```python
print("="*70)
print("üéØ BUSINESS SEGMENT CLASSIFICATION")
print("="*70)

# Define business segments based on RFM scores
def classify_customer_segment(row):
    r, f, m = row['R_Score'], row['F_Score'], row['M_Score']

    # Champions: Best customers (high R, F, M)
    if r >= 4 and f >= 4 and m >= 4:
        return 'Champions'

    # Loyal Customers: High frequency and monetary
    elif f >= 4 and m >= 4:
        return 'Loyal Customers'

    # Potential Loyalists: Recent customers with good frequency
    elif r >= 3 and f >= 3:
        return 'Potential Loyalists'

    # New Customers: Recent but low frequency/monetary
    elif r >= 4 and f <= 2:
        return 'New Customers'

    # At Risk: Were good customers but becoming inactive
    elif r <= 2 and f >= 3 and m >= 3:
        return 'At Risk'

    # Cannot Lose: High monetary but low recency
    elif m >= 4 and r <= 2:
        return 'Cannot Lose Them'

    # Hibernating: Low recency, frequency, but had some value
    elif r <= 2 and f <= 2 and m >= 2:
        return 'Hibernating'

    # Lost Customers: Lowest scores
    elif r <= 2 and f <= 2 and m <= 2:
        return 'Lost'

    # Others: Everything else
    else:
        return 'Need Attention'

# Apply segmentation
df_segment['Customer_Segment'] = df_segment.apply(classify_customer_segment, axis=1)

print("‚úÖ Customer segments assigned!")

# Segment distribution
print("\nüìä CUSTOMER SEGMENT DISTRIBUTION:")
segment_dist = df_segment['Customer_Segment'].value_counts()
segment_dist_pct = (segment_dist / len(df_segment) * 100).round(2)

segment_summary = pd.DataFrame({
    'Segment': segment_dist.index,
    'Customer_Count': segment_dist.values,
    'Percentage': segment_dist_pct.values
}).sort_values('Customer_Count', ascending=False)

display(segment_summary)

# Visualize segment distribution
fig = px.pie(segment_summary, values='Customer_Count', names='Segment',
             title='üéØ Customer Segment Distribution',
             color_discrete_sequence=px.colors.qualitative.Set3,
             hole=0.4)
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.update_layout(height=500)
fig.show()
```

```python
print("="*70)
print("üí∞ SEGMENT-WISE PERFORMANCE METRICS")
print("="*70)

# Key metrics by segment
segment_metrics = df_segment.groupby('Customer_Segment').agg({
    'CLIENTNUM': 'count',
    'Total_Trans_Amt': 'mean',
    'Total_Trans_Ct': 'mean',
    'Credit_Limit': 'mean',
    'Total_Revolving_Bal': 'mean',
    'Avg_Utilization_Ratio': 'mean',
    'Months_Inactive_12_mon': 'mean',
    'Total_Relationship_Count': 'mean',
    'Engagement_Score': 'mean'
}).round(2)

segment_metrics.columns = ['Customer_Count', 'Avg_Trans_Amount', 'Avg_Trans_Count',
                           'Avg_Credit_Limit', 'Avg_Revolving_Bal', 'Avg_Utilization',
                           'Avg_Inactive_Months', 'Avg_Products', 'Avg_Engagement_Score']

# Calculate total revenue contribution per segment
segment_metrics['Total_Revenue'] = (segment_metrics['Customer_Count'] *
                                    segment_metrics['Avg_Trans_Amount']).round(0)

# Sort by total revenue
segment_metrics = segment_metrics.sort_values('Total_Revenue', ascending=False)

print("\nüìä COMPREHENSIVE SEGMENT ANALYSIS:")
display(segment_metrics)

# Revenue contribution visualization
fig = go.Figure()

fig.add_trace(go.Bar(
    x=segment_metrics.index,
    y=segment_metrics['Total_Revenue'],
    marker=dict(color=segment_metrics['Total_Revenue'],
                colorscale='Viridis', showscale=True),
    text=segment_metrics['Total_Revenue'],
    textposition='outside',
    name='Total Revenue'
))

fig.update_layout(
    title='üí∞ Total Revenue Contribution by Customer Segment',
    xaxis_title='Customer Segment',
    yaxis_title='Total Revenue ($)',
    height=500,
    showlegend=False
)
fig.show()
```

```python
print("="*70)
print("‚ö†Ô∏è CHURN ANALYSIS BY CUSTOMER SEGMENT")
print("="*70)

# Churn rate by segment
churn_by_segment = df_segment.groupby('Customer_Segment')['Attrition_Flag'].apply(
    lambda x: (x == 'Attrited Customer').sum() / len(x) * 100
).round(2).sort_values(ascending=False)

churn_counts = df_segment.groupby(['Customer_Segment', 'Attrition_Flag']).size().unstack(fill_value=0)

churn_summary = pd.DataFrame({
    'Segment': churn_by_segment.index,
    'Churn_Rate_%': churn_by_segment.values,
    'Churned_Customers': churn_counts['Attrited Customer'].values,
    'Active_Customers': churn_counts['Existing Customer'].values
})

print("\nüìä CHURN RATE BY SEGMENT:")
display(churn_summary.sort_values('Churn_Rate_%', ascending=False))

# Visualization
fig = make_subplots(
    rows=1, cols=2,
    subplot_titles=('Churn Rate by Segment (%)', 'Churned vs Active Customers'),
    specs=[[{'type': 'bar'}, {'type': 'bar'}]]
)

# Churn rate bar chart
fig.add_trace(
    go.Bar(x=churn_summary['Segment'], y=churn_summary['Churn_Rate_%'],
           marker=dict(color=churn_summary['Churn_Rate_%'],
                      colorscale='Reds', showscale=False),
           text=churn_summary['Churn_Rate_%'],
           textposition='outside',
           name='Churn Rate %'),
    row=1, col=1
)

# Stacked bar for churned vs active
fig.add_trace(
    go.Bar(x=churn_summary['Segment'], y=churn_summary['Active_Customers'],
           name='Active', marker_color='#2ecc71'),
    row=1, col=2
)

fig.add_trace(
    go.Bar(x=churn_summary['Segment'], y=churn_summary['Churned_Customers'],
           name='Churned', marker_color='#e74c3c'),
    row=1, col=2
)

fig.update_layout(height=500, showlegend=True, barmode='stack')
fig.update_xaxes(tickangle=45)
fig.show()

# High-risk segments
print("\nüö® HIGH-RISK SEGMENTS (Churn Rate > 20%):")
high_risk = churn_summary[churn_summary['Churn_Rate_%'] > 20]
if len(high_risk) > 0:
    display(high_risk)
else:
    print("   No segments with churn rate > 20%")
```

```python
print("="*70)
print("üî¨ K-MEANS CLUSTERING VALIDATION")
print("="*70)

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Select features for clustering
clustering_features = ['Total_Trans_Amt', 'Total_Trans_Ct', 'Credit_Limit',
                       'Total_Revolving_Bal', 'Avg_Utilization_Ratio',
                       'Months_Inactive_12_mon', 'Total_Relationship_Count']

X = df_segment[clustering_features].copy()

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Determine optimal number of clusters using elbow method
inertias = []
silhouette_scores = []
K_range = range(3, 9)

from sklearn.metrics import silhouette_score

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    inertias.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(X_scaled, kmeans.labels_))

# Plot elbow curve
fig = make_subplots(rows=1, cols=2,
                    subplot_titles=('Elbow Method', 'Silhouette Score'))

fig.add_trace(
    go.Scatter(x=list(K_range), y=inertias, mode='lines+markers',
               marker=dict(size=10, color='blue'), name='Inertia'),
    row=1, col=1
)

fig.add_trace(
    go.Scatter(x=list(K_range), y=silhouette_scores, mode='lines+markers',
               marker=dict(size=10, color='green'), name='Silhouette Score'),
    row=1, col=2
)

fig.update_xaxes(title_text="Number of Clusters (K)", row=1, col=1)
fig.update_xaxes(title_text="Number of Clusters (K)", row=1, col=2)
fig.update_yaxes(title_text="Inertia", row=1, col=1)
fig.update_yaxes(title_text="Silhouette Score", row=1, col=2)
fig.update_layout(height=400, showlegend=False,
                  title_text="üî¨ Optimal Cluster Analysis")
fig.show()

print(f"\nüìä Silhouette Scores for K=3 to 8:")
for k, score in zip(K_range, silhouette_scores):
    print(f"   K={k}: {score:.4f}")

# Apply optimal clustering (K=5 typically works well)
optimal_k = 5
kmeans_final = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
df_segment['KMeans_Cluster'] = kmeans_final.fit_predict(X_scaled)

print(f"\n‚úÖ K-Means clustering completed with K={optimal_k}")
print(f"   Silhouette Score: {silhouette_score(X_scaled, df_segment['KMeans_Cluster']):.4f}")

# Cluster distribution
print("\nüìä Cluster Distribution:")
display(df_segment['KMeans_Cluster'].value_counts().sort_index())
```

```python
print("="*70)
print("üìã CLUSTER PROFILING & NAMING")
print("="*70)

# Analyze each cluster
cluster_profile = df_segment.groupby('KMeans_Cluster').agg({
    'CLIENTNUM': 'count',
    'Total_Trans_Amt': 'mean',
    'Total_Trans_Ct': 'mean',
    'Credit_Limit': 'mean',
    'Avg_Utilization_Ratio': 'mean',
    'Months_Inactive_12_mon': 'mean',
    'Engagement_Score': 'mean'
}).round(2)

cluster_profile.columns = ['Size', 'Avg_Trans_Amt', 'Avg_Trans_Ct',
                           'Avg_Credit_Limit', 'Avg_Utilization',
                           'Avg_Inactive_Months', 'Avg_Engagement']

print("\nüìä CLUSTER PROFILES:")
display(cluster_profile.sort_values('Avg_Engagement', ascending=False))

# Assign business names to clusters based on characteristics
def name_cluster(cluster_id, profile):
    eng_score = profile.loc[cluster_id, 'Avg_Engagement']
    trans_amt = profile.loc[cluster_id, 'Avg_Trans_Amt']

    if eng_score >= profile['Avg_Engagement'].quantile(0.75):
        return f"Cluster {cluster_id}: Premium High-Value"
    elif eng_score >= profile['Avg_Engagement'].quantile(0.50):
        return f"Cluster {cluster_id}: Growing Mid-Tier"
    elif eng_score >= profile['Avg_Engagement'].quantile(0.25):
        return f"Cluster {cluster_id}: Standard Active"
    else:
        return f"Cluster {cluster_id}: Low Engagement"

cluster_names = {i: name_cluster(i, cluster_profile) for i in range(optimal_k)}
df_segment['Cluster_Name'] = df_segment['KMeans_Cluster'].map(cluster_names)

print("\nüè∑Ô∏è CLUSTER NAMES:")
for k, v in cluster_names.items():
    print(f"   {v}")

# Churn rate by cluster
cluster_churn = df_segment.groupby('Cluster_Name')['Attrition_Flag'].apply(
    lambda x: (x == 'Attrited Customer').sum() / len(x) * 100
).round(2)

print("\n‚ö†Ô∏è CHURN RATE BY CLUSTER:")
display(pd.DataFrame({'Cluster': cluster_churn.index, 'Churn_Rate_%': cluster_churn.values}))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, roc_auc_score

# Label encode target
df_segment['Churn'] = (df_segment['Attrition_Flag'] == 'Attrited Customer').astype(int)

features = ['Customer_Age','Credit_Limit','Total_Trans_Ct','Total_Trans_Amt',
            'Months_Inactive_12_mon','Contacts_Count_12_mon','Avg_Utilization_Ratio',
            'Engagement_Score','R_Score','F_Score','M_Score']

X = df_segment[features]
y = df_segment['Churn']

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.25, random_state=42)

rf = RandomForestClassifier(n_estimators=300, random_state=42, class_weight='balanced')
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)
y_prob = rf.predict_proba(X_test)[:,1]

print(classification_report(y_test, y_pred))
print("ROC-AUC:", roc_auc_score(y_test, y_prob))
imp = pd.Series(rf.feature_importances_, index=features).sort_values(ascending=False)
px.bar(imp, title='üîë Feature Importance - Churn Model').show()
```
