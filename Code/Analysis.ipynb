```python
print("="*70)
print("üîß FEATURE ENGINEERING FOR SEGMENTATION")
print("="*70)

# Create a working copy
df_segment = df.copy()

# RFM Components:
# R (Recency) = Months_Inactive_12_mon (lower is better - more recent activity)
# F (Frequency) = Total_Trans_Ct (higher is better)
# M (Monetary) = Total_Trans_Amt (higher is better)

# Additional behavioral features
df_segment['Avg_Transaction_Value'] = df_segment['Total_Trans_Amt'] / df_segment['Total_Trans_Ct']
df_segment['Credit_Utilization_Pct'] = df_segment['Avg_Utilization_Ratio'] * 100
df_segment['Available_Credit_Pct'] = (df_segment['Avg_Open_To_Buy'] / df_segment['Credit_Limit']) * 100

# Engagement Score (composite metric)
df_segment['Engagement_Score'] = (
    df_segment['Total_Trans_Ct'] * 0.4 +  # Frequency weight
    (df_segment['Total_Trans_Amt'] / 100) * 0.4 +  # Monetary weight (scaled)
    (6 - df_segment['Months_Inactive_12_mon']) * 0.2  # Recency weight (inverted)
)

print("\n‚úÖ New features created:")
print("   ‚Ä¢ Avg_Transaction_Value")
print("   ‚Ä¢ Credit_Utilization_Pct")
print("   ‚Ä¢ Available_Credit_Pct")
print("   ‚Ä¢ Engagement_Score")

print("\nüìä Sample of new features:")
display(df_segment[['Total_Trans_Amt', 'Total_Trans_Ct', 'Avg_Transaction_Value',
                     'Engagement_Score', 'Months_Inactive_12_mon']].head(10))

# Summary statistics
print("\nüìà Engagement Score Distribution:")
print(df_segment['Engagement_Score'].describe())
```

```python
print("="*70)
print("üíØ RFM SCORING SYSTEM")
print("="*70)

# Create RFM scores using quartiles (1-4 scale, 4 being best)
# Using rank-based method to handle duplicates better

# Recency Score (invert: lower months inactive = higher score)
df_segment['R_Score'] = pd.cut(df_segment['Months_Inactive_12_mon'],
                                bins=[-1, 1, 2, 3, 7],
                                labels=[4, 3, 2, 1])

# Frequency Score (higher transaction count = higher score)
df_segment['F_Score'] = pd.qcut(df_segment['Total_Trans_Ct'],
                                 q=4, labels=[1, 2, 3, 4], duplicates='drop')

# Monetary Score (higher transaction amount = higher score)
df_segment['M_Score'] = pd.qcut(df_segment['Total_Trans_Amt'],
                                 q=4, labels=[1, 2, 3, 4], duplicates='drop')

# Convert to numeric (handling any NaN values)
df_segment['R_Score'] = pd.to_numeric(df_segment['R_Score'], errors='coerce').fillna(2).astype(int)
df_segment['F_Score'] = pd.to_numeric(df_segment['F_Score'], errors='coerce').fillna(2).astype(int)
df_segment['M_Score'] = pd.to_numeric(df_segment['M_Score'], errors='coerce').fillna(2).astype(int)

# Combined RFM Score
df_segment['RFM_Score'] = (df_segment['R_Score'] * 100 +
                           df_segment['F_Score'] * 10 +
                           df_segment['M_Score'])

# RFM Segment (string representation)
df_segment['RFM_Segment'] = (df_segment['R_Score'].astype(str) +
                              df_segment['F_Score'].astype(str) +
                              df_segment['M_Score'].astype(str))

print("‚úÖ RFM Scores calculated!")
print("\nüìä Sample RFM Analysis:")
display(df_segment[['CLIENTNUM', 'Months_Inactive_12_mon', 'Total_Trans_Ct',
                     'Total_Trans_Amt', 'R_Score', 'F_Score', 'M_Score',
                     'RFM_Score', 'RFM_Segment']].head(10))

# RFM Score distribution
print("\nüìà RFM Score Distribution:")
print(df_segment['RFM_Score'].describe())

# Top RFM segments
print("\nüèÜ TOP 10 RFM SEGMENTS (by customer count):")
top_segments = df_segment['RFM_Segment'].value_counts().head(10)
display(pd.DataFrame({'RFM_Segment': top_segments.index, 'Customer_Count': top_segments.values}))

# Verify scores are created
print(f"\n‚úÖ R_Score range: {df_segment['R_Score'].min()} to {df_segment['R_Score'].max()}")
print(f"‚úÖ F_Score range: {df_segment['F_Score'].min()} to {df_segment['F_Score'].max()}")
print(f"‚úÖ M_Score range: {df_segment['M_Score'].min()} to {df_segment['M_Score'].max()}")
```

```python
print("="*70)
print("üéØ BUSINESS SEGMENT CLASSIFICATION")
print("="*70)

# Define business segments based on RFM scores
def classify_customer_segment(row):
    r, f, m = row['R_Score'], row['F_Score'], row['M_Score']

    # Champions: Best customers (high R, F, M)
    if r >= 4 and f >= 4 and m >= 4:
        return 'Champions'

    # Loyal Customers: High frequency and monetary
    elif f >= 4 and m >= 4:
        return 'Loyal Customers'

    # Potential Loyalists: Recent customers with good frequency
    elif r >= 3 and f >= 3:
        return 'Potential Loyalists'

    # New Customers: Recent but low frequency/monetary
    elif r >= 4 and f <= 2:
        return 'New Customers'

    # At Risk: Were good customers but becoming inactive
    elif r <= 2 and f >= 3 and m >= 3:
        return 'At Risk'

    # Cannot Lose: High monetary but low recency
    elif m >= 4 and r <= 2:
        return 'Cannot Lose Them'

    # Hibernating: Low recency, frequency, but had some value
    elif r <= 2 and f <= 2 and m >= 2:
        return 'Hibernating'

    # Lost Customers: Lowest scores
    elif r <= 2 and f <= 2 and m <= 2:
        return 'Lost'

    # Others: Everything else
    else:
        return 'Need Attention'

# Apply segmentation
df_segment['Customer_Segment'] = df_segment.apply(classify_customer_segment, axis=1)

print("‚úÖ Customer segments assigned!")

# Segment distribution
print("\nüìä CUSTOMER SEGMENT DISTRIBUTION:")
segment_dist = df_segment['Customer_Segment'].value_counts()
segment_dist_pct = (segment_dist / len(df_segment) * 100).round(2)

segment_summary = pd.DataFrame({
    'Segment': segment_dist.index,
    'Customer_Count': segment_dist.values,
    'Percentage': segment_dist_pct.values
}).sort_values('Customer_Count', ascending=False)

display(segment_summary)

# Visualize segment distribution
fig = px.pie(segment_summary, values='Customer_Count', names='Segment',
             title='üéØ Customer Segment Distribution',
             color_discrete_sequence=px.colors.qualitative.Set3,
             hole=0.4)
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.update_layout(height=500)
fig.show()
```

```python
print("="*70)
print("üî¨ K-MEANS CLUSTERING VALIDATION")
print("="*70)

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Select features for clustering
clustering_features = ['Total_Trans_Amt', 'Total_Trans_Ct', 'Credit_Limit',
                       'Total_Revolving_Bal', 'Avg_Utilization_Ratio',
                       'Months_Inactive_12_mon', 'Total_Relationship_Count']

X = df_segment[clustering_features].copy()

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Determine optimal number of clusters using elbow method
inertias = []
silhouette_scores = []
K_range = range(3, 9)

from sklearn.metrics import silhouette_score

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    inertias.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(X_scaled, kmeans.labels_))

# Plot elbow curve
fig = make_subplots(rows=1, cols=2,
                    subplot_titles=('Elbow Method', 'Silhouette Score'))

fig.add_trace(
    go.Scatter(x=list(K_range), y=inertias, mode='lines+markers',
               marker=dict(size=10, color='blue'), name='Inertia'),
    row=1, col=1
)

fig.add_trace(
    go.Scatter(x=list(K_range), y=silhouette_scores, mode='lines+markers',
               marker=dict(size=10, color='green'), name='Silhouette Score'),
    row=1, col=2
)

fig.update_xaxes(title_text="Number of Clusters (K)", row=1, col=1)
fig.update_xaxes(title_text="Number of Clusters (K)", row=1, col=2)
fig.update_yaxes(title_text="Inertia", row=1, col=1)
fig.update_yaxes(title_text="Silhouette Score", row=1, col=2)
fig.update_layout(height=400, showlegend=False,
                  title_text="üî¨ Optimal Cluster Analysis")
fig.show()

print(f"\nüìä Silhouette Scores for K=3 to 8:")
for k, score in zip(K_range, silhouette_scores):
    print(f"   K={k}: {score:.4f}")

# Apply optimal clustering (K=5 typically works well)
optimal_k = 5
kmeans_final = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
df_segment['KMeans_Cluster'] = kmeans_final.fit_predict(X_scaled)

print(f"\n‚úÖ K-Means clustering completed with K={optimal_k}")
print(f"   Silhouette Score: {silhouette_score(X_scaled, df_segment['KMeans_Cluster']):.4f}")

# Cluster distribution
print("\nüìä Cluster Distribution:")
display(df_segment['KMeans_Cluster'].value_counts().sort_index())
```

```python
print("="*70)
print("üìã CLUSTER PROFILING & NAMING")
print("="*70)

# Re-run K-Means clustering and add KMeans_Cluster if it's missing (to ensure consistency)
# This is added for robustness against inconsistent kernel states or out-of-order execution.
if 'KMeans_Cluster' not in df_segment.columns:
    from sklearn.preprocessing import StandardScaler
    from sklearn.cluster import KMeans

    clustering_features = ['Total_Trans_Amt', 'Total_Trans_Ct', 'Credit_Limit',
                           'Total_Revolving_Bal', 'Avg_Utilization_Ratio',
                           'Months_Inactive_12_mon', 'Total_Relationship_Count']

    X = df_segment[clustering_features].copy()
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    optimal_k = 5 # Using the previously determined optimal_k
    kmeans_final = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
    df_segment['KMeans_Cluster'] = kmeans_final.fit_predict(X_scaled)
    print("‚úÖ K-Means clustering re-executed to ensure 'KMeans_Cluster' is present.")

# Analyze each cluster
cluster_profile = df_segment.groupby('KMeans_Cluster').agg({
    'CLIENTNUM': 'count',
    'Total_Trans_Amt': 'mean',
    'Total_Trans_Ct': 'mean',
    'Credit_Limit': 'mean',
    'Avg_Utilization_Ratio': 'mean',
    'Months_Inactive_12_mon': 'mean',
    'Engagement_Score': 'mean'
}).round(2)

cluster_profile.columns = ['Size', 'Avg_Trans_Amt', 'Avg_Trans_Ct',
                           'Avg_Credit_Limit', 'Avg_Utilization',
                           'Avg_Inactive_Months', 'Avg_Engagement']

print("\nüìä CLUSTER PROFILES:")
display(cluster_profile.sort_values('Avg_Engagement', ascending=False))

# Assign business names to clusters based on characteristics
def name_cluster(cluster_id, profile):
    eng_score = profile.loc[cluster_id, 'Avg_Engagement']
    trans_amt = profile.loc[cluster_id, 'Avg_Trans_Amt']

    if eng_score >= profile['Avg_Engagement'].quantile(0.75):
        return f"Cluster {cluster_id}: Premium High-Value"
    elif eng_score >= profile['Avg_Engagement'].quantile(0.50):
        return f"Cluster {cluster_id}: Growing Mid-Tier"
    elif eng_score >= profile['Avg_Engagement'].quantile(0.25):
        return f"Cluster {cluster_id}: Standard Active"
    else:
        return f"Cluster {cluster_id}: Low Engagement"

# Ensure optimal_k is defined for mapping
if 'optimal_k' not in locals():
    optimal_k = 5 # Defaulting to 5 if not previously defined

cluster_names = {i: name_cluster(i, cluster_profile) for i in range(optimal_k)}
df_segment['Cluster_Name'] = df_segment['KMeans_Cluster'].map(cluster_names)

print("\n‚ùÇÔ∏è CLUSTER NAMES:")
for k, v in cluster_names.items():
    print(f"   {v}")

# Churn rate by cluster
cluster_churn = df_segment.groupby('Cluster_Name')['Attrition_Flag'].apply(
    lambda x: (x == 'Attrited Customer').sum() / len(x) * 100
).round(2)

print("\n‚ö†Ô∏è CHURN RATE BY CLUSTER:")
display(pd.DataFrame({'Cluster': cluster_churn.index, 'Churn_Rate_%': cluster_churn.values}))
```

```python



import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score, roc_curve, confusion_matrix
import plotly.express as px
import plotly.graph_objects as go

# Load dataset
df = pd.read_csv('BankChurners.csv')
# Feature Engineering
df['Engagement_Score'] = (
    df['Total_Trans_Ct'] * 0.4 +
    (df['Total_Trans_Amt'] / 100) * 0.4 +
    (6 - df['Months_Inactive_12_mon']) * 0.2
)

df['R_Score'] = pd.cut(df['Months_Inactive_12_mon'], bins=[-1, 1, 2, 3, 7], labels=[4, 3, 2, 1]).astype(int)
df['F_Score'] = pd.qcut(df['Total_Trans_Ct'], q=4, labels=[1, 2, 3, 4])
df['M_Score'] = pd.qcut(df['Total_Trans_Amt'], q=4, labels=[1, 2, 3, 4])
df[['F_Score', 'M_Score']] = df[['F_Score', 'M_Score']].apply(pd.to_numeric)
df['Churn'] = (df['Attrition_Flag'] == 'Attrited Customer').astype(int)

# Features and target
features = [
    'Customer_Age', 'Credit_Limit', 'Total_Trans_Ct', 'Total_Trans_Amt',
    'Months_Inactive_12_mon', 'Contacts_Count_12_mon', 'Avg_Utilization_Ratio',
    'Engagement_Score', 'R_Score', 'F_Score', 'M_Score'
]
X = df[features]
y = df['Churn']

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.25, random_state=42)

# Train Random Forest
rf = RandomForestClassifier(n_estimators=300, random_state=42, class_weight='balanced')
rf.fit(X_train, y_train)

# Predictions
y_prob = rf.predict_proba(X_test)[:, 1]
y_pred = (y_prob > 0.5).astype(int)

# ROC curve
fpr, tpr, thresholds = roc_curve(y_test, y_prob)
roc_auc = roc_auc_score(y_test, y_prob)

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
cm_df = pd.DataFrame(cm, index=['Actual No Churn', 'Actual Churn'], columns=['Predicted No Churn', 'Predicted Churn'])

# Add churn probability to test set
df_test = X_test.copy()
df_test['Churn_Prob'] = y_prob
df_test['Predicted_Churn'] = y_pred

# ROC Curve Plot
fig_roc = go.Figure()
fig_roc.add_trace(go.Scatter(x=fpr, y=tpr, mode='lines', name='ROC Curve'))
fig_roc.add_trace(go.Scatter(x=[0, 1], y=[0, 1], mode='lines', name='Random Guess', line=dict(dash='dash')))
fig_roc.update_layout(title=f"ROC Curve (AUC = {roc_auc:.3f})", xaxis_title="False Positive Rate", yaxis_title="True Positive Rate")
fig_roc.show()

# Churn probability heatmap
fig_prob = px.histogram(df_test, x='Churn_Prob', color='Predicted_Churn',
                        nbins=30, title="üìä Churn Probability Distribution",
                        labels={'Churn_Prob': 'Predicted Churn Probability'},
                        color_discrete_sequence=['#2ecc71', '#e74c3c'])
fig_prob.show()

roc_auc
```

```python
print("="*70)
print("üíé CUSTOMER LIFETIME VALUE (CLV) CALCULATION")
print("="*70)

# Create CLV working dataframe from the most recent 'df' which has RFM scores and Engagement_Score
df_clv = df.copy()

# Define the classify_customer_segment function again to ensure it's available
def classify_customer_segment(row):
    r, f, m = row['R_Score'], row['F_Score'], row['M_Score']

    # Champions: Best customers (high R, F, M)
    if r >= 4 and f >= 4 and m >= 4:
        return 'Champions'

    # Loyal Customers: High frequency and monetary
    elif f >= 4 and m >= 4:
        return 'Loyal Customers'

    # Potential Loyalists: Recent customers with good frequency
    elif r >= 3 and f >= 3:
        return 'Potential Loyalists'

    # New Customers: Recent but low frequency/monetary
    elif r >= 4 and f <= 2:
        return 'New Customers'

    # At Risk: Were good customers but becoming inactive
    elif r <= 2 and f >= 3 and m >= 3:
        return 'At Risk'

    # Cannot Lose: High monetary but low recency
    elif m >= 4 and r <= 2:
        return 'Cannot Lose Them'

    # Hibernating: Low recency, frequency, but had some value
    elif r <= 2 and f <= 2 and m >= 2:
        return 'Hibernating'

    # Lost Customers: Lowest scores
    elif r <= 2 and f <= 2 and m <= 2:
        return 'Lost'

    # Others: Everything else
    else:
        return 'Need Attention'

# Apply segmentation to df_clv
df_clv['Customer_Segment'] = df_clv.apply(classify_customer_segment, axis=1)

# Calculate Customer Lifetime (in months)
df_clv['Customer_Lifetime_Months'] = df_clv['Months_on_book']

# Calculate Average Monthly Transaction Value
df_clv['Monthly_Trans_Value'] = df_clv['Total_Trans_Amt'] / 12  # Annualized

# Calculate Monthly Transaction Frequency
df_clv['Monthly_Trans_Frequency'] = df_clv['Total_Trans_Ct'] / 12

# Estimate Monthly Revenue (simplified model)
# Assuming bank earns 2.5% interchange fee + interest from revolving balance
INTERCHANGE_FEE_RATE = 0.025  # 2.5% on transactions
ANNUAL_INTEREST_RATE = 0.18   # 18% APR on revolving balance
MONTHLY_INTEREST_RATE = ANNUAL_INTEREST_RATE / 12

df_clv['Monthly_Revenue_Interchange'] = df_clv['Monthly_Trans_Value'] * INTERCHANGE_FEE_RATE
df_clv['Monthly_Revenue_Interest'] = df_clv['Total_Revolving_Bal'] * MONTHLY_INTEREST_RATE
df_clv['Total_Monthly_Revenue'] = (df_clv['Monthly_Revenue_Interchange'] +
                                    df_clv['Monthly_Revenue_Interest'])

# Calculate Customer Lifetime Value
# CLV = Monthly Revenue √ó Customer Lifetime (months) √ó Retention Rate
# We'll use segment-specific retention rates

# Calculate retention rate by segment (1 - churn_rate)
segment_retention = df_clv.groupby('Customer_Segment')['Attrition_Flag'].apply(
    lambda x: 1 - ((x == 'Attrited Customer').sum() / len(x))
)

df_clv['Retention_Rate'] = df_clv['Customer_Segment'].map(segment_retention)

# Calculate CLV (simplified: Monthly Revenue √ó Expected Lifetime)
# Expected Lifetime = Customer_Lifetime_Months √ó Retention_Rate
df_clv['Expected_Lifetime_Months'] = df_clv['Customer_Lifetime_Months'] * df_clv['Retention_Rate']
df_clv['Customer_Lifetime_Value'] = df_clv['Total_Monthly_Revenue'] * df_clv['Expected_Lifetime_Months']

print("‚úÖ CLV calculated using:")
print("   ‚Ä¢ Interchange fee revenue (2.5% on transactions)")
print("   ‚Ä¢ Interest revenue (18% APR on revolving balance)")
print("   ‚Ä¢ Segment-specific retention rates")
print("   ‚Ä¢ Customer lifetime adjusted by retention")

print("\nüìä CLV Summary Statistics:")
print(df_clv['Customer_Lifetime_Value'].describe())

print("\nüí∞ Average CLV by Segment:")
clv_by_segment = df_clv.groupby('Customer_Segment')['Customer_Lifetime_Value'].mean().sort_values(ascending=False)
display(pd.DataFrame({
    'Segment': clv_by_segment.index,
    'Avg_CLV ($)': clv_by_segment.values.round(2)
}))

# Visualize CLV distribution
fig = px.box(df_clv, x='Customer_Segment', y='Customer_Lifetime_Value',
             color='Customer_Segment',
             title='Customer Lifetime Value Distribution by Segment',
             labels={'Customer_Lifetime_Value': 'CLV ($)', 'Customer_Segment': 'Segment'})
fig.update_layout(showlegend=False, height=500)
fig.update_xaxes(tickangle=45)
fig.show()
```

```python
print("="*70)
print("üèÜ PROFITABILITY TIER CLASSIFICATION")
print("="*70)

# Define profitability tiers based on CLV
df_clv['CLV_Percentile'] = pd.qcut(df_clv['Customer_Lifetime_Value'],
                                     q=4, labels=['Low', 'Medium', 'High', 'Premium'],
                                     duplicates='drop')

# More granular profitability classification
def classify_profitability(clv_value):
    if clv_value >= df_clv['Customer_Lifetime_Value'].quantile(0.90):
        return 'Platinum (Top 10%)'
    elif clv_value >= df_clv['Customer_Lifetime_Value'].quantile(0.75):
        return 'Gold (Top 25%)'
    elif clv_value >= df_clv['Customer_Lifetime_Value'].quantile(0.50):
        return 'Silver (Top 50%)'
    elif clv_value >= df_clv['Customer_Lifetime_Value'].quantile(0.25):
        return 'Bronze (Top 75%)'
    else:
        return 'Standard'

df_clv['Profitability_Tier'] = df_clv['Customer_Lifetime_Value'].apply(classify_profitability)

print("‚úÖ Profitability tiers assigned!")

# Tier distribution
print("\nüìä PROFITABILITY TIER DISTRIBUTION:")
tier_dist = df_clv['Profitability_Tier'].value_counts()
display(pd.DataFrame({
    'Tier': tier_dist.index,
    'Customer_Count': tier_dist.values,
    'Percentage': (tier_dist.values / len(df_clv) * 100).round(2)
}))

# CLV by profitability tier
print("\nüí∞ AVERAGE METRICS BY PROFITABILITY TIER:")
tier_metrics = df_clv.groupby('Profitability_Tier').agg({
    'Customer_Lifetime_Value': 'mean',
    'Total_Trans_Amt': 'mean',
    'Total_Monthly_Revenue': 'mean',
    'Credit_Limit': 'mean',
    'Retention_Rate': 'mean'
}).round(2)

tier_metrics.columns = ['Avg_CLV', 'Avg_Trans_Amt', 'Avg_Monthly_Revenue',
                        'Avg_Credit_Limit', 'Avg_Retention_Rate']
tier_metrics = tier_metrics.sort_values('Avg_CLV', ascending=False)
display(tier_metrics)

# Visualize
fig = px.sunburst(df_clv,
                  path=['Profitability_Tier', 'Customer_Segment'],
                  values='Customer_Lifetime_Value',
                  title='Profitability Hierarchy: Tier ‚Üí Segment ‚Üí CLV',
                  color='Customer_Lifetime_Value',
                  color_continuous_scale='Viridis')
fig.update_layout(height=600)
fig.show()
```

```python
print("="*70)
print("üíµ REVENUE & COST ANALYSIS")
print("="*70)

# Estimate customer acquisition cost (CAC) and servicing cost
AVG_ACQUISITION_COST = 150  # Industry average for credit cards
MONTHLY_SERVICING_COST = 5  # Operational cost per customer per month

df_clv['Acquisition_Cost'] = AVG_ACQUISITION_COST
df_clv['Total_Servicing_Cost'] = MONTHLY_SERVICING_COST * df_clv['Customer_Lifetime_Months']
df_clv['Total_Cost'] = df_clv['Acquisition_Cost'] + df_clv['Total_Servicing_Cost']

# Calculate Net Profit
df_clv['Total_Revenue'] = df_clv['Total_Monthly_Revenue'] * df_clv['Customer_Lifetime_Months']
df_clv['Net_Profit'] = df_clv['Total_Revenue'] - df_clv['Total_Cost']
df_clv['Profit_Margin_%'] = (df_clv['Net_Profit'] / df_clv['Total_Revenue'] * 100).round(2)

# ROI (Return on Investment)
df_clv['ROI_%'] = ((df_clv['Net_Profit'] / df_clv['Total_Cost']) * 100).round(2)

print("‚úÖ Cost & profitability metrics calculated!")
print(f"   ‚Ä¢ Acquisition Cost: ${AVG_ACQUISITION_COST}")
print(f"   ‚Ä¢ Monthly Servicing Cost: ${MONTHLY_SERVICING_COST}")

print("\nüìä PROFITABILITY SUMMARY:")
profitability_summary = pd.DataFrame({
    'Metric': ['Total Revenue', 'Total Cost', 'Net Profit', 'Avg Profit Margin (%)', 'Avg ROI (%)'],
    'Value': [
        f"${df_clv['Total_Revenue'].sum():,.0f}",
        f"${df_clv['Total_Cost'].sum():,.0f}",
        f"${df_clv['Net_Profit'].sum():,.0f}",
        f"{df_clv['Profit_Margin_%'].mean():.2f}%",
        f"{df_clv['ROI_%'].mean():.2f}%"
    ]
})
display(profitability_summary)

# Profitability by segment
print("\nüí∞ NET PROFIT BY CUSTOMER SEGMENT:")
segment_profit = df_clv.groupby('Customer_Segment').agg({
    'Total_Revenue': 'sum',
    'Total_Cost': 'sum',
    'Net_Profit': 'sum',
    'ROI_%': 'mean'
}).round(0)

segment_profit = segment_profit.sort_values('Net_Profit', ascending=False)
display(segment_profit)

# Visualize profit by segment
fig = go.Figure()

segments = segment_profit.index
fig.add_trace(go.Bar(name='Revenue', x=segments, y=segment_profit['Total_Revenue'],
                     marker_color='#2ecc71'))
fig.add_trace(go.Bar(name='Cost', x=segments, y=segment_profit['Total_Cost'],
                     marker_color='#e74c3c'))
fig.add_trace(go.Bar(name='Net Profit', x=segments, y=segment_profit['Net_Profit'],
                     marker_color='#3498db'))

fig.update_layout(
    title='Revenue, Cost & Net Profit by Customer Segment',
    xaxis_title='Customer Segment',
    yaxis_title='Amount ($)',
    barmode='group',
    height=500
)
fig.update_xaxes(tickangle=45)
fig.show()
```
